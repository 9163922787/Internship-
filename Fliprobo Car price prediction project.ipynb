```python
from selenium import webdriver
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
url='https://www.cars24.com/'
driver.get(url)
time.sleep(10)


#close the popup for location choice
try:
    close = driver.find_element_by_xpath('//div[@class="float-right TBxAN"]/img')
    close.click()
except NoSuchElementException:
    print('Could not find popup for location')

time.sleep(2)
#Click the city list

try:
    city = driver.find_element_by_xpath('//p[@class="_1OR7d"]')
    city.click()
except NoSuchElementException:
    print('Could not find specified location for city')
time.sleep(5)

#Search for city
try:
    select_city = driver.find_element_by_xpath('//ul[@class="_16Bvy"]/li[1]/span/img')
    select_city.click()
except NoSuchElementException:
    print('Could not find specified location for city')
time.sleep(5)

#click the used car
try:
    close = driver.find_element_by_xpath('//ul[@class="float-right _1yq3e"]/li[3]/a')
    close.click()
except NoSuchElementException:
    print('Could not find popup for location')

time.sleep(5)

#declaring arrays
cars24_year=[]
cars24_brand=[]
cars24_varient=[]
cars24_model=[]
cars24_type=[]
cars24_price=[]
cars24_meter=[]
cars24_owner=[]
cars24_fuel=[]

SCROLL_PAUSE_TIME = 3
CYCLES = 20

#scrolling the page till end
html = driver.find_element_by_tag_name('html')
html.send_keys(Keys.PAGE_DOWN)  
html.send_keys(Keys.PAGE_DOWN)  
time.sleep(SCROLL_PAUSE_TIME * 3)

for j in range(CYCLES):
    html.send_keys(Keys.END)
    time.sleep(SCROLL_PAUSE_TIME)

# scraping the year of car, brand, varient    
try:
    name=driver.find_elements_by_xpath('//h2[@class="_3FpCg"]')
    for i in name:
        cars24_year.append(i.text.split(" ")[0])
        cars24_brand.append(i.text.split(" ")[1])
        cars24_varient.append(i.text.replace('[','').replace(']','').split(" ")[2:])
except NoSuchElementException:
    print('Could not find brand, year ,varient')
    
# scraping the model   
try:
    model=driver.find_elements_by_xpath('//p[@class="cvakB"]')
    for i in name:
        cars24_model.append(i.text)
except NoSuchElementException:
    print('Could not find model')
    
#scraping the type automatic or manual 
try:
    cartype=driver.find_elements_by_xpath('//p[@class="cvakB"]/span')
    for i in cartype:
        cars24_type.append(i.text)
except NoSuchElementException:
    print('Could not find type of car')
    

#scraping the price of car
try:
    price=driver.find_elements_by_xpath('//div[@class="_7udZZ"]/span')
    for i in price:
        cars24_price.append(i.text.replace('â‚¹','').replace(',',''))
except NoSuchElementException:
    print('Could not find price')

#scraping the kms of car
try:
    price=driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[1]')
    for i in price:
        cars24_meter.append(i.text)
except NoSuchElementException:
    print('Could not kms ')

    
# scraping the owner 1st or 2nd or 3rd
try:
    owner=driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[2]')
    for i in owner:
        cars24_owner.append(i.text)
except NoSuchElementException:
    print('Could not find 1st or 2nd owner')

# scraping the type of fuel
try:
    fuel=driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[3]')
    for i in fuel:
        cars24_fuel.append(i.text)
except NoSuchElementException:
    print('Could not find fuel type')

# creating dictionary for first 500 data
dict={"manufactured":cars24_year[0:500],"brand":cars24_brand[0:500],"varient":cars24_varient[0:500],"owner":cars24_owner[0:500],"fuel":cars24_fuel[0:500],"type":cars24_type[0:500],"kms":cars24_meter[0:500],"price":cars24_price[0:500]}
# creating dataframe of data
df1=pd.DataFrame(dict)
# scaping the city of the car
c_name = driver.find_elements_by_xpath('//p[@class="_1OR7d"]/label')
# adding the city column in data 
for i in c_name:
    city_name=i.text
df1['city']=city_name
df1
```

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
url='https://www.cardekho.com/'
driver.get(url)
time.sleep(10)

#link for used car
try:
    used_cars = driver.find_element_by_xpath("/html/body/div[2]/div/header/div[2]/div/div/nav/ul/li[2]/ul/li[2]/a")  
except NoSuchElementException:
    print('link for used car not found.')
time.sleep(2)
used_cars = used_cars.get_attribute('href')
driver.get(used_cars)
time.sleep(3)
try:
    cities = driver.find_elements_by_xpath("//ul[@class = 'listing gsc_col-md-9 gsc_col-lg-9']//li/a")
except NoSuchElementException:
    print('Xpath of ')

time.sleep(10)
city = []
for i in cities:
    city.append(i.get_attribute('href'))
time.sleep(10)

cdekho_year=[]
cdekho_brand=[]
cdekho_varient=[]
cdekho_type=[]
cdekho_price=[]
cdekho_owner=[]
cdekho_meter=[]
cdekho_fuel=[]
cdekho_urls=[]
cdekho_city=[]
for j in city:
    driver.get(j)
    time.sleep(3)
    
    for _ in range(10):
        time.sleep(3)
        driver.execute_script("window.scrollBy(0,500)")
        
    current_url = driver.current_url    
    try:
        car = driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
    except NoSuchElementException:
        print('link for used car not found.')

    for i in car:
        cdekho_urls.append(i.get_attribute('href'))
        
    #gsc_col-xs-7  carsName
    try:
        for i in driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a"):
            cdekho_year.append(i.text.split(" ")[0])
            cdekho_brand.append(i.text.split(" ")[1])
            cdekho_varient.append(i.text.split(" ")[2:])        
    except NoSuchElementException:
        cdekho_year.append('-')
        cdekho_brand.append('-')
        cdekho_varient.append('-')     
        print('Brand not found.')
   
    try:
        #amnt 
        for i in driver.find_elements_by_xpath("//span[@class='amnt ']"):
            cdekho_price.append(i.text)
    except NoSuchElementException:
        cdekho_price.append('-')
        print('amount not found.')
   
    try:
        #Kms runned
        for i in driver.find_elements_by_xpath("//div[@class='truncate dotlist']//span[1]"):
            cdekho_meter.append(i.text)
    except NoSuchElementException:
        cdekho_meter.append('-')
        print('KMS runned not found.')

    try:
        #Fuel type
        for i in driver.find_elements_by_xpath("//div[@class='truncate dotlist']//span[2]"):
            cdekho_fuel.append(i.text)
    except NoSuchElementException:
        cdekho_fuel.append('-')
        print('fuel type not found.')

    try:
        #manual or automatic type
        for i in driver.find_elements_by_xpath("//div[@class='truncate dotlist']//span[3]"):
            cdekho_type.append(i.text)
            cdekho_city.append(j.split('+')[-1])
    except NoSuchElementException:
        cdekho_type.append('-')
        cdekho_city.append('-')
        print('transfer type not found.')

    
time.sleep(10)

for i in cdekho_urls:
    driver.get(i)
    time.sleep(2)
    #No Of Owner(s)
    #fetching owner
    try:
        #owner = driver.find_element_by_xpath("//*[text()='Owner']/following-sibling::div")
        owner = driver.find_element_by_xpath("//ul[@class='gsc_row clearfix']/li[4]/div/div")
        cdekho_owner.append(owner.text)
    except NoSuchElementException:
        cdekho_owner.append('-')

#dict={"manufactured":cdekho_year[0:300],"brand":cdekho_brand[0:300],"varient":cdekho_varient[0:300],"owner":cdekho_owner[0:300],"fuel":cdekho_fuel[0:300],"type":cdekho_type[0:300],"kms":cdekho_meter[0:300],"price":cdekho_price[0:300],"city":cdekho_city[0:300]}

#frame=pd.DataFrame(dict)
#frame
```
