```python
# loading the driver and opening the website
driver=webdriver.Chrome(executable_path="C:\\Users\\User\\chromedriver.exe")
#driver.maximize_window()
url='https://www.flipkart.com/'
driver.get(url)
time.sleep(5)
```

```python
def scrap_data(rating_urls):    
    rating=[]
    comment=[]
    review=[]
    # looping though saved page url of reviews
    for i in rating_urls:
        driver.get(i)
        time.sleep(5)
        #scraping the rating
        try:
            rat=driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']")
            for i in rat:
                if(i.text==''):
                    rating.append('-')
                else:
                    rating.append(i.text)
        except NoSuchElementException:
            print("Could not find ratings")
        time.sleep(1.5)
        #scraping the comment
        try:
            cmmnt=driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
            for i in cmmnt:
                if(i.text==''):
                    comment.append('-')
                else:
                    comment.append(i.text)
        except NoSuchElementException:
            print("Could not find Comment")
        time.sleep(1.5)
        #scraping the review
        try:
            rvw=driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div")
            for i in rvw:
                if(i.text==''):
                    review.append('-')
                else:
                    review.append(i.text)
        except NoSuchElementException:
            print("Could not find Review")
        time.sleep(2)
    
    m=min(len(rating),len(comment),len(review))
    dict1={'Rating':rating[:m],'Comment':comment[:m],'Review':review[:m]}
    return dict1
```

```python
rating_urls=[]
# loop through list items
for x in items:
    # code for sending the items list wise in search input and click
    try:
        search_input=driver.find_element_by_xpath("//input[@class='_3704LK']")
        search_input.clear()
        search_input.send_keys(x)

        search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
        search_btn.click()
        time.sleep(7)
    except NoSucheElementException:
        print('Could not enter the item in search button')
    
    # saving the link url of each items of category
    links=[]
    
    try:
        item_href=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
        for i in item_href:
            links.append(i.get_attribute('href'))
    except NoSuchElementException:
        print('Could find the links for the items')
    time.sleep(2.5)  
    #loop through the links for each item link
    for i in links:
        driver.get(i)
        time.sleep(6)
        #click to view all reviews
        try:
            all_review=driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']/span")
            all_review.click()
        except NoSuchElementException:
            print("Could not click the all_review",x)
        
        time.sleep(3) 
        # saving the url of pagination for all the reviews
        
        try:
            page=driver.find_elements_by_xpath("//nav[@class='yFHi8N']/a")
            for i in page:
                rating_urls.append(i.get_attribute('href'))

        except NoSuchElementException:
            print("Could not find and paginationb")
        
            
```

```python
# loading the driver and opening the website
driver=webdriver.Chrome(executable_path="C:\\Users\\User\\chromedriver.exe")
#driver.maximize_window()
url='https://www.amazon.in/'
driver.get(url)
time.sleep(5)
```

```python
srch_items=['laptops','smart phones','headphones','smart watchs','Professional Cameras','Printers','Monitors']
```

```python
title = []
review_text = []
ratings = []
```

```python
for i in srch_items:
    srchbar = driver.find_element_by_id("twotabsearchtextbox")
    srchbar.send_keys(i)
    driver.find_element_by_id("nav-search-submit-button").click()  #clicking on search button
    time.sleep(3)
    item_url = []
    for i in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
        item_url.append(i.get_attribute('href'))
    for i in item_url:
        driver.get(i)
        time.sleep(1)
        for _ in range(2):
            driver.execute_script("window.scrollBy(0,6000)")
            time.sleep(3)
        #click on see all reviews
        try:
            driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']").click()
        except:
            pass
        
        url_ = []
        try:
            two_str = driver.find_element_by_xpath("//table[@id='histogramTable']/tbody/tr[4]/td[2]/a")
            url_.append(two_str.get_attribute('href'))
        except: pass
        try:
            three_str = driver.find_element_by_xpath("//table[@id='histogramTable']/tbody/tr[3]/td[2]/a")
            url_.append(three_str.get_attribute('href'))
        except: pass
        try:
            one_str = driver.find_element_by_xpath("//table[@id='histogramTable']/tbody/tr[5]/td[2]/a")
            url_.append(one_str.get_attribute('href'))
        except: pass
        try:
            fiv_str = driver.find_element_by_xpath("//table[@id='histogramTable']/tbody/tr[1]/td[2]/a")
            url_.append(fiv_str.get_attribute('href'))
        except: pass
        try:
            four_str = driver.find_element_by_xpath("//table[@id='histogramTable']/tbody/tr[2]/td[2]/a")
            url_.append(four_str.get_attribute('href'))
        except: pass
        
        for i in url_:
            driver.get(i)
            time.sleep(5)
            for i in range(3):
                ids = driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none review-views celwidget']/div")
                comment_ids = []
                for i in ids:
                    comment_ids.append(i.get_attribute('id'))

                for x in comment_ids:
                    #Extract user ids from each user on a page
                    try:
                        review_title = driver.find_element_by_xpath('//*[@id="' + x +'"]/div/div/div[2]/a[2]/span[1]')
                        title.append(review_title.get_attribute("innerHTML"))
                    except NoSuchElementException:
                        title.append('')

                    try:
                        rating = driver.find_element_by_xpath('//*[@id="' + x +'"]/div/div/div[2]/a[1]/i/span[1]')
                        ratings.append(rating.get_attribute("innerHTML"))
                    except NoSuchElementException:
                        ratings.append('')

                    try:
                        text = driver.find_element_by_xpath('//*[@id="' + x +'"]/div/div/div[4]/span/span')
                        review_text.append(text.get_attribute("innerHTML"))
                    except NoSuchElementException:
                        review_text.append('')
                try:
                    driver.find_element_by_xpath("//ul[@class='a-pagination']/li[2]/a").click()
                    time.sleep(5)
                except : break 
```

```python
print(len(title))
print(len(ratings))
print(len(review_text))
```

# Data cleaning for rating list

```python
ratings_copy=ratings
```

```python
ratings
```

```python
#spliting the rating list to get only star rating number
star_rating=[]
for i in ratings_copy:
    i=i.split('.')[0]
    star_rating.append(i)
```

```python
#replacin the null value with '-'
for i in range(len(star_rating)):
    if len(star_rating[i])==0:
        star_rating[i]='-'
```

```python
len(star_rating)
```

```python
#saving in a dictionary
dict1={'Rating':star_rating,'Comment':title,'Review':review_text}
```

```python
#appending the extracted data to existing dataframe
scraped_data=pd.DataFrame(dict1)
df=df.append(scraped_data)
```
