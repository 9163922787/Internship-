```python
# Function to search 300 images

def scrap_image(item):
    #  Enter x in “Search” field 
    
    count=0 # To keep count of number of images (upto 300)
    
    driver=webdriver.Firefox(executable_path="C:\\Users\\User\\geckodriver.exe")
    # get the webpage www.amajon.com
    driver.get('https://www.amajon.in/')
    time.sleep(5)    
    search=driver.find_element_by_id("twotabsearchtextbox")
    search.send_keys(item)

    # click the search button

    search_buton=driver.find_element_by_id("nav-search-submit-button")
    search_buton.click()
    time.sleep(5)    
    images=[]    
    while (count<=300):
        elements=driver.find_elements_by_xpath("//img[@class ='s-image' ]")
        count=count+len(elements)
        
        for i in elements:
            try:
                image=i.get_attribute('src')
                images.append(image)
            except NoSuchElementException:
                images.append('-')
        
        if (count<=300):                    
            # Clicking on next page
            driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']").click()
            
            time.sleep(5)
    driver.close()       
    return(images)
            
```

```python
sharee_images=scrap_image('Sharees')[:300]
len(sharee_images)
```

```python
trousers_images=scrap_image('Trouser for men')[:300]
len(trousers_images)
```

```python
jeans_images=scrap_image('Jeans for men')[:300]
len(jeans_images)
```

```python
# Creating Directories
import os

def directory(dir):
    current_path=os.getcwd()
    new=os.path.join(current_path,dir)
    if not os.path.exists(new):
        os.makedirs(new)
        
directory('Sharees_women')
directory('Trousers_men')
directory('Jeans_men')

# Dowloading images

import shutil
import requests

for index, link in enumerate(sharee_images):
    print('Downloading {0} of 300 saree images'.format(index+1))
    response=requests.get(link)
    with open('Sharees_women/img{0}.jpg'.format(index+1),"wb") as file:
        file.write(response.content)
        
for index, link in enumerate(trousers_images):
    print('Downloading {0} of 300 trouser images'.format(index+1))
    response=requests.get(link)
    with open('Trousers_men/img{0}.jpg'.format(index+1),"wb") as file:
        file.write(response.content)
        
for index, link in enumerate(jeans_images):
    print('Downloading {0} of 300 jeans images'.format(index+1))
    response=requests.get(link)
    with open('Jeans_men/img{0}.jpg'.format(index+1),"wb") as file:
        file.write(response.content)
```
